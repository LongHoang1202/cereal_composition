```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cereal_df= pd.read_csv('C:/Users/ADMINH/Desktop/python for data analytic/cereal.csv')
cereal_df.head()#tên, nsx, loại nóng lạnh, shelf: kệ để hàng
```

```python
fields= ['shelf','weight','cups','rating']
```

```python
new_cereal = cereal_df.drop(fields, axis=1)#loại bỏ các cột ko liên quan
new_cereal
```

```python
#tính corelation các giá trị
cereal_corr = new_cereal.corr()
cereal_corr2= cereal_corr.iloc[1:,:-1]
cereal_corr2
```

```python
#sử dụng ones_like để tạo ma trận boolean
ones_corr= np.ones_like(cereal_corr, dtype= bool)
ones_corr
```

```python
#sử dụng hàm triu để chỉ trả về nửa trên của matrix corr
mask= np.triu(ones_corr)
mask2=mask[1:,:-1] #bỏ dòng đầu , cột cuối
```

```python
fig, ax= plt.subplots(figsize=(8,6))

#cài đặt màu cho biểu đồ
cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)

sns.heatmap(data= cereal_corr2, mask=mask2,
            annot=True, fmt=".2f",
            cmap=cmap,
            vmin=-1, vmax=1,
            linewidths=0.5) 
#annot để hiện giá trị trong ô, fmt để rút gọn số thập phân, vmin vimax để color hiện từ -1,1

ytick= [i.upper() for i in cereal_corr2.index]
xtick= [i.upper() for i in cereal_corr2.columns]
#viết hoa cho các chữ

ax.set_yticklabels(ytick, rotation=0, fontsize=10);
ax.set_xticklabels(xtick, rotation=90, fontsize=10);
tittle="CORRELATION MATRIX\nSAMPLED CEREALS COMPOSITION\n"#dấu \ để cách dòng cho tên biểu đồ
ax.set_title(tittle, loc= 'left', fontsize=18);
```
